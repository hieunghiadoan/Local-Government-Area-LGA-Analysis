```python
import pandas as pd
import numpy as np
```

```python
students_csv = "/Users/nghiadoanhieu/Monash_Bootcamp/Homework/Module_Challenge/pandas-challenge/Resources/students_complete.csv"
schools_csv = "/Users/nghiadoanhieu/Monash_Bootcamp/Homework/Module_Challenge/pandas-challenge/Resources/schools_complete.csv"
```

```python
students_df = pd.read_csv(students_csv)
schools_df = pd.read_csv(schools_csv)
students_df
```

```python
merged_df = students_df.merge(schools_df, how='left', on='school_name')
```

```python
#Local Government Area (LGA) Summary
unique_school_names = schools_df['school_name'].unique()
school_count = len(unique_school_names)

total_students = schools_df['size'].sum()
total_budget = schools_df['budget'].sum()

#number and % passing reading
passing_reading = students_df.loc[students_df['reading_score'] >= 70]['reading_score'].count()
perc_pass_reading = (passing_reading/total_students) * 100
#number and % passing math
passing_math = students_df.loc[students_df['maths_score'] >= 70]['maths_score'].count()
perc_pass_math = (passing_math/total_students) * 100
#average scores
avg_math_score = students_df['maths_score'].mean()
avg_reading_score = students_df['reading_score'].mean()
overall_passing = (perc_pass_math + perc_pass_reading) / 2

district_summary = pd.DataFrame({
    
    "Total Schools": [school_count],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Reading Score": [avg_reading_score],
    "Average Math Score": [avg_math_score],
    "% Passing Reading":[perc_pass_reading],
    "% Passing Math": [perc_pass_math],
    "Overall Passing Rate": [overall_passing]

})
district_summary
```

```python
#School Summary
by_school = merged_df.set_index('school_name').groupby(['school_name'])
sch_types = schools_df.set_index('school_name')['type']
stu_per_sch = by_school['Student ID'].count()

sch_budget = schools_df.set_index('school_name')['budget']
stu_budget = schools_df.set_index('school_name')['budget']/schools_df.set_index('school_name')['size']

avg_math = by_school['maths_score'].mean()
avg_read = by_school['reading_score'].mean()

# % passing
pass_math = merged_df[merged_df['maths_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch * 100
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch * 100
overall = (pass_math + pass_read) / 2
school_summary = pd.DataFrame({
    "School Type": sch_types,
    "Total Students": stu_per_sch,
    "Per Student Budget": stu_budget,
    "Total School Budget": sch_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
})
school_summary
```

```python
#Top Performing Schools By Passing Rate
top_schools = school_summary.sort_values("Overall Passing Rate", ascending = False)
top_schools.head()
```

```python
#Bottom Performing Schools By Pass Rate
bottom_schools = school_summary.sort_values("Overall Passing Rate")
bottom_schools.head()
```

```python
#Math Scores By Year
ninth_year_math = students_df.loc[students_df['year'] == '9'].groupby('school_name')["maths_score"].mean()
tenth_year_math = students_df.loc[students_df['year'] == '10'].groupby('school_name')["maths_score"].mean()
eleventh_year_math = students_df.loc[students_df['year'] == '11'].groupby('school_name')["maths_score"].mean()
twelfth_year_math = students_df.loc[students_df['year'] == '12'].groupby('school_name')["maths_score"].mean()

maths_scores = pd.DataFrame({
        "9": ninth_year_math,
        "10": tenth_year_math,
        "11": eleventh_year_math,
        "12": twelfth_year_math
})
maths_scores.head(10)
```

```python
#Reading Scores By Year
ninth_year_reading = students_df.loc[students_df['year'] == '9'].groupby('school_name')["reading_score"].mean()
tenth_year_reading = students_df.loc[students_df['year'] == '10'].groupby('school_name')["reading_score"].mean()
eleventh_year_reading = students_df.loc[students_df['year'] == '11'].groupby('school_name')["reading_score"].mean()
twelfth_year_reading = students_df.loc[students_df['year'] == '12'].groupby('school_name')["reading_score"].mean()

reading_scores_by_year = pd.DataFrame({
        "9": ninth_year_reading,
        "10": tenth_year_reading,
        "11": eleventh_year_reading,
        "12": twelfth_year_reading
})
reading_scores_by_year
```

```python
#Scores by school spending
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
merged_df['budget_bins'] = pd.cut(merged_df['budget']/merged_df['size'], spending_bins, labels)
school_spending_df = merged_df.groupby('budget_bins')

spending_maths_scores = school_spending_df["maths_score"].mean()
spending_reading_scores = school_spending_df["reading_score"].mean()
spending_passing_maths = merged_df[merged_df["maths_score"] >= 70].groupby("budget_bins")["Student ID"].count() / school_spending_df["Student ID"].count() * 100
spending_passing_reading = merged_df[merged_df["reading_score"] >= 70].groupby("budget_bins")["Student ID"].count() / school_spending_df["Student ID"].count() * 100
overall_passing_spending = merged_df[(merged_df["reading_score"] >= 70) & (merged_df["maths_score"] >= 70)].groupby("budget_bins")["Student ID"].count() / school_spending_df["Student ID"].count() * 100

                     
school_spending_df = pd.DataFrame({
    "Average Math Score": spending_maths_scores,
    "Average Reading Score": spending_reading_scores,
    '% Passing Math': spending_passing_maths,
    '% Passing Reading': spending_passing_reading,
    "Overall Passing Rate":overall_passing_spending 
            
})
school_spending_df 
```

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]


district_summary['School Size'] = pd.cut(district_summary['Total Students'], size_bins, labels=labels)
by_size = district_summary.groupby('School Size')
avg_math = by_size['Average Math Score'].mean()
avg_read = by_size['Average Reading Score'].mean()
pass_math = by_size['% Passing Math'].mean()
pass_read = by_size['% Passing Reading'].mean()
overall_pass = by_size['Overall Passing Rate'].mean()


size_summary = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall_pass
})
size_summary.head(10)
```

```python
#Scores by school type
group_type = merged_df.groupby("type")

avg_math = group_type['maths_score'].mean()
avg_read = group_type['maths_score'].mean()
pass_math = merged_df[merged_df['maths_score'] >= 70].groupby('type')['Student ID'].count()/group_type['Student ID'].count() * 100
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('type')['Student ID'].count()/group_type['Student ID'].count() * 100
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['maths_score'] >= 70)].groupby('type')['Student ID'].count()/group_type['Student ID'].count() * 100
           
scores_by_type = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall}) 
scores_by_type
```
